```python
import numpy as np
import gensim
import json
```

```python
eng_alph = set('abcdefghijklmnopqrstuvwxyz')
```

```python
config = json.load(open(r"C:\Users\szymon\Desktop\pycharm\babel\scripts\load_config.json", "r"))
kv = gensim.models.KeyedVectors.load_word2vec_format(config["data_path"], binary=config["data_path"].endswith('.bin'))
```

```python
wordlist = list(kv.key_to_index.keys())
```

```python
a = np.zeros((26, 26))
```

```python
for w in wordlist:
    w = w.lower()
    if set(w).issubset(eng_alph):
        for i in range(1, len(w)):
            a[ord(w[i - 1]) - ord('a')][ord(w[i]) - ord('a')] += 1
```

```python
a
```

```python
probabilities = a / np.reshape(np.sum(a, axis=-1), (26, 1
                                                ))
```
